```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
#Add column which contains only the year
propertiesM.insert(3, 'Year', propertiesM.Yr_Month.dt.year)
```

```python
#Add column which identifies each 'Borough' as a Region, Borough, or the entire country
propertiesM.loc[(propertiesM.Borough.str.isupper()), 'Area_Type']='Region'
propertiesM.loc[(propertiesM.Area_Type.isnull()),'Area_Type'] ='Borough'
propertiesM.loc[(propertiesM.Borough=='England'), 'Area_Type']='Country'
```

```python
#Visualize the trends around the UK, London is the most expensive and growing the fastest 
regions = propertiesM[propertiesM.Area_Type =='Region'].groupby(['Borough','Year'], as_index ='False').last()
regions.pivot_table(index ='Year', columns ='Borough', values ='Average_Price').plot(figsize =(18,12))
plt.title('Average Price of a House in Regions of the UK from 1995-2020', fontsize = 20)
plt.ylabel('Average Price', fontsize =16)
plt.xlabel('Year', fontsize =16)
plt.show()
```

```python
#create summarized df on a yearly basis 
average_price = propertiesM.loc[:,['Year','Borough', 'Average_Price','Area_Type']]
```

```python
#use the 'last' average price of the year (December)
average_price = average_price.groupby(['Borough','Year','Area_Type'], as_index = False).last()
```

```python
#calculate the percent change year over year for one borough 
Hack_Price = average_price[average_price.Borough =='Hackney']
hack_percent = average_price.loc[average_price.Borough =='Hackney',['Year','Average_Price']].set_index(['Year']).pct_change()*100
hack_percent.reset_index(inplace = True)
Hack_Price = pd.merge(Hack_Price, hack_percent, on ='Year')
Hack_Price.columns=['Borough', 'Year', 'Area_Type', 'Average_Price', 'Pct_Change']
```

```python
#function to calculate the % change in housing prices from 1999 - 2019 
def create_price_ratio(borough):
    avg_1999 = df.loc[(df.Borough ==borough) &(df.Year==1999),'Average_Price'].values[0]
    avg_2019 = df.loc[(df.Borough ==borough) &(df.Year==2019),'Average_Price'].values[0]
    global ratio 
    ratio = ((avg_2019 - avg_1999)/avg_2019)
    return ratio
```
